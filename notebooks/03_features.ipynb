combine (merge) both datasets to prepare one clean table for modeling.

```python
import pandas as pd

df_energy = pd.read_csv("../data/bronze/maroc_energie_mix_1990_2023.csv")
df_weather = pd.read_csv("../data/bronze/weather_maroc_1990_2023.csv")

df_final = df_energy.merge(df_weather, on="annee", how="left")
df_final.to_csv("../data/silver/maroc_energy_weather_1990_2023.csv", index=False)

print(df_final.head())
```
